```python
from transformers import pipeline

# Load model GPT-2 untuk text generation
generator = pipeline('text-generation', model='gpt2')
```

```python
def generate_response(prompt):
    # Generate jawaban menggunakan model GPT-2
    response = generator(prompt, max_length=100, num_return_sequences=1)
    return response[0]['generated_text']
```

```python
import gradio as gr

def chat_interface(text):
    return generate_response(text)

# Buat antarmuka Gradio
iface = gr.Interface(
    fn=chat_interface,  # Fungsi yang akan dipanggil
    inputs="text",      # Input berupa teks
    outputs="text",     # Output berupa teks
    title="Bot Materi Bangun Ruang SD",  # Judul antarmuka
    description="Masukkan pertanyaan Anda tentang materi bangun ruang SD."
)

# Jalankan antarmuka
iface.launch()
```

```python
def load_material(file_path):
    with open(file_path, 'r') as file:
        material = file.read()
    return material

# Contoh penggunaan
material = load_material('materi_bangun_ruang.txt')
print(material)
```

```python
def generate_response(prompt):
    # Gabungkan prompt dengan materi dari file
    full_prompt = f"{material}\n\nPertanyaan: {prompt}\nJawaban:"
    response = generator(full_prompt, max_length=150, num_return_sequences=1)
    return response[0]['generated_text']
```
